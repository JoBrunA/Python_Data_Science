### Importación de datos



```python
import pandas as pd


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print(ingreso_tienda1)#1150880400.0
print(ingreso_tienda2)#1116343500.0
print(ingreso_tienda3)#1098019600.0
print(ingreso_tienda4)#1038375700.0
```

# 2. Ventas por categoría

```python
# Tienda 1
ventas_tienda = tienda["Categoría del Producto"].value_counts()
print("Tienda 1 - Productos vendidos por categoría:")
print(ventas_tienda)
print("\n")

# Tienda 2
ventas_tienda2 = tienda2["Categoría del Producto"].value_counts()
print("Tienda 2 - Productos vendidos por categoría:")
print(ventas_tienda2)
print("\n")
```

```python
# Tienda 3
ventas_tienda3 = tienda3["Categoría del Producto"].value_counts()
print("Tienda 3 - Productos vendidos por categoría:")
print(ventas_tienda3)
print("\n")

# Tienda 4
ventas_tienda4 = tienda4["Categoría del Producto"].value_counts()
print("Tienda 4 - Productos vendidos por categoría:")
print(ventas_tienda4)
print("\n")
```

# 3. Calificación promedio de la tienda


```python
# ---- 1. Calificación promedio general por tienda ----
print(f"Tienda 1: {tienda['Calificación'].mean():.2f}")
print(f"Tienda 2: {tienda2['Calificación'].mean():.2f}")
print(f"Tienda 3: {tienda3['Calificación'].mean():.2f}")
print(f"Tienda 4: {tienda4['Calificación'].mean():.2f}")
```

```python
# ---- 2. Calificación promedio por producto dentro de cada tienda ----
print("Tienda 1")
print(tienda.groupby("Producto")["Calificación"].mean().sort_values(ascending=False).head(3))
print("Tienda 2")
print(tienda2.groupby("Producto")["Calificación"].mean().sort_values(ascending=False).head(3))
print("Tienda 3")
print(tienda3.groupby("Producto")["Calificación"].mean().sort_values(ascending=False).head(3))
print("Tienda 4")
print(tienda4.groupby("Producto")["Calificación"].mean().sort_values(ascending=False).head(3))
```

# 4. Productos más y menos vendidos

```python
#!pip install matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


def mostrar_mas_menos(df, nombre_tienda):
    ventas = df["Producto"].value_counts()    
    print(f"\n{nombre_tienda} - Producto más vendido:")
    print(ventas.head(3))

    print(f"{nombre_tienda} - Producto menos vendido:")
    print(ventas.tail(3))

    ventas.head(3).plot(kind="bar", title=f"{nombre_tienda} - Top 3 Productos Más Vendidos", color='lightgreen')
    plt.ylabel("Cantidad vendida")
    plt.xlabel("Producto")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
# Aplicar la función a cada tienda
mostrar_mas_menos(tienda, "Tienda 1")
mostrar_mas_menos(tienda2, "Tienda 2")
mostrar_mas_menos(tienda3, "Tienda 3")
mostrar_mas_menos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular costo de envío promedio por tienda
promedio_envio_tienda1 = tienda["Costo de envío"].mean()
promedio_envio_tienda2 = tienda2["Costo de envío"].mean()
promedio_envio_tienda3 = tienda3["Costo de envío"].mean()
promedio_envio_tienda4 = tienda4["Costo de envío"].mean()

print(f"Costo de envío promedio - Tienda 1: ${promedio_envio_tienda1:.2f}")
print(f"Costo de envío promedio - Tienda 2: ${promedio_envio_tienda2:.2f}")
print(f"Costo de envío promedio - Tienda 3: ${promedio_envio_tienda3:.2f}")
print(f"Costo de envío promedio - Tienda 4: ${promedio_envio_tienda4:.2f}")
```

```python
# Preparar datos para el gráfico
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
costos = [promedio_envio_tienda1, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]

# Crear gráfico de líneas
plt.figure(figsize=(8,5))
plt.plot(tiendas, costos, marker='o', linestyle='-', color='purple')
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo de envío promedio")
plt.grid(True)

# Mostrar el valor exacto en cada punto
for i, costo in enumerate(costos):
    plt.text(i, costo + 0.1, f"${costo:.2f}", ha='center', color='black')

plt.show()
```
